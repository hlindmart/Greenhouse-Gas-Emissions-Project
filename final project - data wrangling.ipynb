```python
# defining function for rename 
def reName (series):
    if series=="carbon_dioxide_co2_emissions_without_land_use_land_use_change_and_forestry_lulucf_in_kilotonne_co2_equivalent":
        return "CO2 emissions wo LULUCF"
    if series=="greenhouse_gas_ghgs_emissions_without_land_use_land_use_change_and_forestry_lulucf_in_kilotonne_co2_equivalent":
        return "GHG emissions wo LULUCF"
    if series=="methane_ch4_emissions_without_land_use_land_use_change_and_forestry_lulucf_in_kilotonne_co2_equivalent":
        return "Methane emissions wo LULUCF"
    if series=="nitrous_oxide_n2o_emissions_without_land_use_land_use_change_and_forestry_lulucf_in_kilotonne_co2_equivalent":
        return "Nitrous oxide emissions wo LULUCF"
    if series=="sulphur_hexafluoride_sf6_emissions_in_kilotonne_co2_equivalent":
        return "Sulfur emissions"
    if series=="hydrofluorocarbons_hfcs_emissions_in_kilotonne_co2_equivalent":
        return "HCFS emissions"
    if series=="greenhouse_gas_ghgs_emissions_including_indirect_co2_without_lulucf_in_kilotonne_co2_equivalent":
        return "GHG emissions including CO2 wo LULUCF"
    if series=="perfluorocarbons_pfcs_emissions_in_kilotonne_co2_equivalent":
        return "PFCS emissions"
    if series=="nitrogen_trifluoride_nf3_emissions_in_kilotonne_co2_equivalent":
        return "NF3 emissions"
    if series=="unspecified_mix_of_hydrofluorocarbons_hfcs_and_perfluorocarbons_pfcs_emissions_in_kilotonne_co2_equivalent":
        return "Unspecified mix of HCFS and PFCS emissions"
```

```python
# defining function for recode
def recode (series):
    if series=="CO2 emissions wo LULUCF":
        return 0
    if series=="GHG emissions wo LULUCF":
        return 1
    if series=="Methane emissions wo LULUCF":
        return 2
    if series=="Nitrous oxide emissions wo LULUCF":
        return 3
    if series=="Sulfur emissions":
        return 4
    if series=="HCFS emissions":
        return 5
    if series=="GHG emissions including CO2 wo LULUCF":
        return 6
    if series=="PFCS emissions":
        return 7
    if series=="NF3 emissions":
        return 8
    if series=="Unspecified mix of HCFS and PFCS emissions":
        return 9
```

```python
# attaching reName to data
ghgInv['category']=ghgInv['category'].apply(reName)
```

```python
# attaching recode to data
ghgInv['categoryR']=ghgInv['category'].apply(recode)
```

```python
# dropping column mistakenly created
ghgInv.drop('catRename', axis=1, inplace=True)
```
